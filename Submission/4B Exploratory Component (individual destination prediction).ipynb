```python
import pandas as pd
```

```python
df = pd.read_csv("NYC_Weather_2016_2022.csv")


df
df['time'] = pd.to_datetime(df['time'])
```

```python
holidays = [
    ("2018-01-01", "New Year's Day"),
    ("2018-01-15", "Martin Luther King Jr. Day"),
    ("2018-02-12", "Lincoln's Birthday (NY State)"),
    ("2018-02-19", "Presidents' Day / Washington's Birthday"),
    ("2018-05-28", "Memorial Day"),
    ("2018-07-04", "Independence Day"),
    ("2018-09-03", "Labor Day"),
    ("2018-10-08", "Columbus Day"),
    ("2018-11-06", "Election Day (NY State)"),
    ("2018-11-12", "Veterans Day (Observed)"),
    ("2018-11-22", "Thanksgiving Day"),
    ("2018-12-25", "Christmas Day")
]

df_holidays = pd.DataFrame(holidays, columns=["time", "holiday_name"])
df_holidays
```

```python
df_hourly = df[df['time'].dt.year == 2018].copy()
df_hourly.reset_index(drop=True, inplace=True)


# Ensure original df['time'] is datetime (done earlier)
df['time'] = pd.to_datetime(df['time'], errors='coerce')
df = df.dropna(subset=['time'])  # optional: drop unparsable rows

# Filter for 2018 (preserves full datetime in 'time')
df_hourly = df[df['time'].dt.year == 2018].copy()
df_hourly.reset_index(drop=True, inplace=True)

# Create a new date-only column for merging (preserve full datetime in `time`)
df_hourly['date'] = df_hourly['time'].dt.date

# Prepare holidays with a matching 'date' column
df_holidays['date'] = pd.to_datetime(df_holidays['time']).dt.date

# Merge on 'date' (not 'time')
df_merged = df_hourly.merge(df_holidays[['date','holiday_name']], on='date', how='left')

# Create is_holiday flag
df_merged['is_holiday'] = df_merged['holiday_name'].notna().astype(int)

# If you want to keep both the original timestamp and the date column:
# df_merged[['time','date','holiday_name','is_holiday']].head()
df_merged
#download df_merged as csv
df_merged.to_csv("nyc_weather_2018_with_holidays.csv", index=False)
```
